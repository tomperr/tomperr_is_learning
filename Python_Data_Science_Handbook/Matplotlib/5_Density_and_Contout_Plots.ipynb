# Density and Contour Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

## Visualizing a Three-Dimensional Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

A contour plot can be created with the `plt.contour` function. The easiest way to prepare the data is to use `np.meshgrid`

```python
x = np.linspace(0, 5, 500)
y = np.linspace(0, 5, 400)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

By default, when a single color is used, negative values are reprensented by dashed lines, and positive values by solid lines. Colors can be specified with `cmap` parameter

```python
plt.contour(X, Y, Z, 20, cmap='RdGy'); # RdGy stands for Red-Gray
```

All possible colors:

```python
dir(plt.cm)
```

Instead of simple lines, we can fill spaces with `plt.contourf()`

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

The problem now is that colors is discrete rather than continuous. A way to solve the problem is to use `plt.imshow()`

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
plt.axis("image")
```

Sometimes, it could be useful to combine `plt.contour` and `plt.imshow`. For example, to have a transparent background, with contours and labels over. 

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
