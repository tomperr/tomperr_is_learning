# Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Let's start by creating a figure and axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
ax.plot(x, np.cos(x));
```

The equivalent with the `pylab` interface is

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

Use the `color` and `linestyle` parameters of `plot()`

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

Combination of both

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Adjusting the Plot: Axes Limits

Matplotlib does it automatically, but it can also be controlled with `xlim` and `ylim` methods

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

With reversed bounds, functions can be "inversed"

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

A useful related method is `plt.axis()`, that takes limits as `[xmin, xmax, ymin, ymax]`

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

`axis` has also "presets" : `tight`, `equal`...

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labelling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

When multiple lines are drawn, we usually use `plt.legend()`

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## Aside: Matplotlib Gotchas

For some methods, the name are some the same for both interfaces

- `plt.xlabel()` -> `ax.set_xlabel()`
- `plt.ylabel()` -> `ax.set_ylabel()`
- `plt.xlim()` -> `ax.set_xlim()`
- `plt.ylim()` -> `ax.set_ylim()`
- `plt.title()` -> `ax.set_title()`

With the object-oriented interface, it's more common to use `ax.set()`

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
